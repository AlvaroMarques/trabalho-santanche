## Conversa entre Zumbis

Considere que zumbis podem conversar entre si através de um método especial `pergunta`, este método recebe uma pergunta (`String`) e retorna uma resposta (`String`). Podem ser feitas perguntas como a idade, nome e estado. Por exemplo, o método pode receber a pergunta:
~~~
"Qual a sua idade?"
~~~

e então retorna a idade.

Implemente este método nas três classes (`GarotoZumbi`, `Monstro` e `Perfeito`).

```java
public class GarotoZumbi
{
    protected int idade;
    protected String estado;
    protected String nome;
    
    public GarotoZumbi(int idade, String estado, String nome) {
        this.idade = idade;
        this.estado = estado;
        this.nome = nome;
    }
    
    public void mostra() {
        // cabeleira
        if (idade >= 2)
            System.out.println("  *");
        
        // corpo com olhos
        if (estado.equalsIgnoreCase("acordado"))
            System.out.println(" o*o");
        else
            System.out.println(" -*-");
        
        // barba
        if (idade >= 3)
            System.out.println("*****");
        
        System.out.println(nome);
        
        System.out.println();
    }
    
    public void cresce() {
        if (idade < 3)
            idade++;
        mostra();
    }
    
    public void acorda() {
        estado = "acordado";
        mostra();
    }
    
    public void dorme() {
        estado = "dormindo";
        mostra();
    }
    
    public String pergunta(String perg){
        String resposta;
        System.out.print("["+nome+"] ");
        if (perg.equalsIgnoreCase("Qual é a sua idade?")){
            System.out.println("Tenho " + Integer.toString(idade) + ((idade == 1)?" ano":" anos"));
            return "I"+idade;
        }else if (perg.equalsIgnoreCase("Qual o seu nome?")){
            System.out.println("Me chamo " + nome);
            return "N"+nome;
        }else{
            System.out.println("Estou " + estado);
            return "E"+estado;
        }
    }
}

GarotoZumbi garoto = new GarotoZumbi(1, "acordado", "Asdrubal");
garoto.mostra();
garoto.dorme();
garoto.cresce();
garoto.pergunta("qual é a sua idade?");
garoto.pergunta("qual o seu nome?");
garoto.pergunta("qual o seu estado?");
```

```java
public class Monstro extends GarotoZumbi{
    Monstro(int idade, String estado, String nome){
        super(idade, estado, nome);
    }
    
    public void mostraMonstro(){
        if (idade >= 2)
            System.out.println("  *");
        
        // corpo com olhos
        if (estado.equalsIgnoreCase("acordado"))
            System.out.println(" o*o");
        else
            System.out.println(" -*-");
        
        // barba
        if (idade >= 3)
            System.out.println("*****");

        if (idade >= 4)
            System.out.println("#####");
        if (idade  == 5)
            System.out.println("/   \\");
    }
    public void cresceMonstro(){
        if (idade < 5) idade++;
    }
        
}
```

```java
public class Perfeito extends GarotoZumbi{
    private static final String estados[] = new String[]{"normal", "milionario","doido"};
    private int estadoh = 0;
    Perfeito(int idade, String estado, String nome){
        super(idade, estado, nome);
        switch (estado){
            case "normal":
                estadoh = 0;
                break;
            case "milionario":
                estadoh = 1;
                break;
            case "doido":
                estadoh = 2;
                break;
            default:
                estadoh = -1;
                break;
        }
    }
    
    @Override
    public void mostra(){
        mostraPerfeito();
    }
    
    public void mudaEstado(){
        estadoh = ++estadoh%3;
        estado = estados[estadoh];
        
    }
    
    public void mostraPerfeito(){
        System.out.println("+-----+");
        if (idade > 1){
            System.out.println("|     |");
        }
        if (estadoh == 0){
            System.out.println("| o-o |");
            System.out.println("| ___ |");
        }else if (estadoh == 1){
            System.out.println("| $-$ |");
            System.out.println("| ___ |");
        }else{ 
            System.out.println("| @-@ |");
            System.out.println("|  ~  |");
        }
        if (idade < 3){
            System.out.println(" \\___/");
        }else{
            System.out.println("|     |\n\\_____/");
        }
        
    }
    
}
Perfeito garoto = new Perfeito(1, "doido", "Asdrubal");
garoto.mostra();
//garoto.mudaEstado();
garoto.cresce();
garoto.pergunta("qual é a sua idade?");
garoto.pergunta("qual o seu nome?");
garoto.pergunta("qual o seu estado?");
```

## Engajando zumbis na conversa

Escreva um método que permita *engajar* dois objetos zumbi em uma conversa. Após engajar dois zumbis, um deles deve ser capaz de contactar o outro para fazer perguntas.

Construa uma classe zumbi entrevistador que faça perguntas a outros zumbis para selecionar o mais velho entre eles. Se dois ou mais zumbis têm a mesma idade é dado preferência ao primeiro.

Escreva um programa em que um zumbi entrevistador entreviste três zumbis candidatos. Ele deve imprimir uma mensagem indicando o nome do mais velho entre os zumbis. Você deve *engajar*  cada um dos candidatos com o entrevistador, que informará o resultado quando o método `resultado` for acionado.

```java
public class ZumbiEntrevistador extends GarotoZumbi{
    ZumbiEntrevistador(int idade, String estado, String nome){
        super(idade, estado, nome);
        
    }
    public void engajar(GarotoZumbi g1, GarotoZumbi g2, GarotoZumbi g3){
        int maxIdade = -1;
        GarotoZumbi current = g1;
        int idadeGaroto;
        GarotoZumbi gzs[] = {g1,g2,g3};
        System.out.println("Qual a idade de vocês?");
        for (GarotoZumbi g : gzs){
            idadeGaroto = Integer.parseInt(g.pergunta("qual é a sua idade?").substring(1));

            if (idadeGaroto > maxIdade){
                current = g;
                maxIdade = idadeGaroto;
            }      
        }
        System.out.println("O mais velho é "+ current.nome);
    }
}
```
